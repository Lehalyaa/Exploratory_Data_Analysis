```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    'Days' : ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours' : [2, 3, 2.3, 4, 3.5],
    'Marks' : [70, 80, 85, 55, 90],
    'Attendance' : [60, 70, 65, 80, 75],
    'Sleep_Hours': [6,7,8,7,6],
    'Assignments_Completed' : [1, 2, 1, 3, 2]
}

df = pd.DataFrame(data)
df
```

```python
# 1.Line Chart (Trend)

plt.plot(df['Days'], df['Study_Hours'], marker = 'o')

plt.title('Study Hours Trend')
plt.xlabel('Days')
plt.ylabel('Study Hours')

plt.show()
```

```python
# Advanced Line Chart

plt.figure(figsize = (7,7))

plt.plot(df['Days'] , df['Study_Hours'], marker = 'o', color = 'blue', label = 'Study_Hours')

plt.plot(df['Days'] , df['Marks'], marker = 's', color = 'red', label = 'Marks')

plt.title('Study Hours and Marks Trend')
plt.xlabel('Days')
plt.ylabel('Study Hours/Marks')
plt.legend()

plt.show()
```

```python
# 2. Bar Chart

plt.bar(df['Days'], df['Marks'], color = 'violet')

plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')

plt.show()
```

```python
# Advanced Bar Graph

bars = plt.bar(df['Days'], df['Marks'], color = 'bisque')

for bar in bars:
  y = bar.get_height()
  plt.text(bar.get_x() + bar.get_width() / 2, y, str(y), ha = 'center', va = 'bottom')

plt.title("Marks Comparison")
plt.xlabel("Days")
plt.ylabel("Marks")

plt.grid(axis = 'y') # Adds Horizontal grid for readability

plt.show()
```

```python
# 3.Hisogram (distribution)

# shows how the values are spread

data = {
    'Marks' : [50, 60, 55, 70, 65, 75, 80, 85, 90, 45, 68, 72, 77, 82, 88, 67, 90, 95, 67]
}

df = pd.DataFrame(data)

plt.hist(df['Marks'] , bins = 15, color = 'paleturquoise')
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
# 4. Scatter Plot (Relationship)

import matplotlib.pyplot as plt

plt.scatter(df['Study_Hours'], df['Marks'], color = 'slateblue')

plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.show()
```

```python
# Conditional Scatter Plot
#Example Category

df['Result'] = ['Pass', 'Fail', 'Pass', 'Fail', 'Fail']

colors = ['red' if r == 'Fail' else 'green' for r in df['Result']]
plt.scatter(df['Study_Hours'], df['Marks'], c = colors)

plt.title("Study Hours vs Marks (Result-wise)")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.show()
```
