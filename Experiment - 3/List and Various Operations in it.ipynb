```python
# Access Items

example = ["abc", 34, True, 40, "male"]
print(example[2])
print(example[4])
```

```python
# Reverse / Negative Indexing

example = ["abc", 34, True, 40, "male"]
print(example[-1])
print(example[-3])
print(example[-5])
```

```python
# List Slicing

subjects = ["Physics", "Chemistry", "Maths", "English", "Hindi", "Computer Science"]
print(subjects[1:4])
print(subjects[-5:-1])# Reverse Indexing
print(subjects[:4]) # To print without the starting index
print(subjects[2:]) # To print without the ending index
```

```python
# Change Item Value

subjects = ["Physics", "Chemistry", "Maths", "English", "Hindi", "Computer Science"]
subjects[-2] = "PE"
subjects[0] = "IT"
print(subjects)
```

```python
# Problem Statement - 1 - Slicing Techniques

numbers = [10,20,30,40,50]
print("Elements from index 1 to 3")
print(numbers[1:4], "\n")

print("First Three Elements")
print(numbers[:3], "\n")

print("Elements from Index 2 onwards")
print(numbers[2:], "\n")

print("All Elements")
print(numbers[:] , "\n")
```

```python
#Problem Statement - 2 - Student Marks Analysis

student_marks = [80 , 88 , 95, 99 , 96, 75, 94, 90, 78, 69]
print("Highest Mark - ", max(student_marks))
print("Lowest Mark - ", min(student_marks))
print("Average Mark - ", sum(student_marks)/len(student_marks))
student_marks.sort()
print("Sorted Marks:", student_marks)
```

```python
# Problem Statement - 3 - Grocery Analysis

grocery = ["Dal", "Oil", "Soap", "Masala", "Vegetables"]
grocery_1 = ["Tooth Paste", "Tooth Brush", "Biscuits", "Detergent"]
print(grocery, "\n")
print("The List wtih extra added Items: ")
grocery.extend(grocery_1)
print(grocery, "\n")
print("The list with removed Items: ")
grocery.remove("Masala")
print(grocery, "\n")
print("The Final Grocery List")
print(grocery)
```

```python
# Problem Statement - 4 - Contact Analysis

contact = ["Aditi Mishra", "Gayatri Bhosle", "Lehalyaa", "Dev Anand", "Ayush Dumbre"]
contact.append("Akshara Sree")
print("The Updated List is: ", contact, "\n")
contact.remove("Ayush Dumbre")
print("The Removed List is: ", contact, "\n")
contact.sort()
print("The sorted List is: ", contact, "\n")
```

```python
# Problem Statement - 5 - Temperature Analysis

temp = [ 34, 45, 38, 37, 32 , 29, 30 , 28, 31 , 30]
print("The first 5 days Temperature are: ", temp[0:5], "\n")
print("The last 5 days Temperature are: ", temp[-5:], "\n")
print("The highest Temperature: ", max(temp), "\n")
print("The lowest Temperature: ", min(temp), "\n")
print("The average Temperature: ", sum(temp)/len(temp), "\n")
```
