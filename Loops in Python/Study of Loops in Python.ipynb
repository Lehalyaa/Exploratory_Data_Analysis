```python
''' Problem Statement - 8   -   Write a Python program to calculate the gross salary
of an employee based on basic salary.'''

basic_sal = int(input("Enter the Basic Salary: "))

if(basic_sal <= 10000):
  HRA = (20/100) * basic_sal
  DA = (80/100) * basic_sal

elif(basic_sal <= 20000):
  HRA = (25/100) * basic_sal
  DA = (90/100) * basic_sal

else:
  HRA = (30/100) * basic_sal
  DA = (95/100) * basic_sal

gross_sal = basic_sal + HRA + DA

print("The Gross Salary is: ", gross_sal)

```

```python
''' Problem Statement - 9   -   Write a Python program to calculate the income
tax based on annual income'''

salary = int(input("Enter the salary: "))

if(salary <= 250000):
  tax = 0
elif(salary <= 500000):
  tax = (salary - 250000) * 0.05

elif(salary <= 1000000):
  tax = (250000 * 0.05) + (500000 * 0.20) + (salary - 1000000) * 0.30

print("The Tax amount to be paid is ", tax)
print("The salary In - Hand is ", salary - tax)
```
