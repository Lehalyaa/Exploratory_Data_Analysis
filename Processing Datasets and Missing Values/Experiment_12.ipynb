## Data Preprocessing of Dataset - 1

```python
data = {
    "Roll No" : [1, 2, 3, 4, 5, 6, 7, 8, 9],
    "Name" : ["Akash", "Bejoy", "Christi", "Dev", "Emily", "Farhan", "Gautham", "Harini", "Iliz"],
    "Age" : [20, "-", 19, 20, 21, 20, "-", 18, "22"],
    "Marks" : ["90", 95, "-", 85, 65, "-", 89, 88, 98],
    "Department" : ["CSE", "ENTC", "CSE", "AIML", "RA", "ENTC", "entc", "CSE", "CSE"],
    "Admission Date" : ["22-09-2026", "01-08-2026", "05-08-2026", "22-09-2026", "01-08-2026", "05-08-2026", "26/09/2026", "22-09-2026", "01-08-2026"]
}

df = pd.DataFrame(data)

# Save as CSV
df.to_csv("Student_Data.csv", index = False)
```

```python
df.replace("-", np.nan, inplace = True)
df
```

```python
# Step 2: Convert columns to numeric where neeeded

df["Age"] = pd.to_numeric(df["Age"], errors = 'coerce')
df["Marks"] = pd.to_numeric(df["Marks"], 'coerce')
df
```

```python
# Step 3: Check Missing values

print("\n Missing Values:")
print(df.isnull().sum())
```

```python
# fill missing age with mean

df["Age"].fillna(df["Age"].mean(), inplace = True)
df

# Fill missing Marks with Median

'''for vareity data median  is used'''
df["Marks"].fillna(df["Marks"].median(), inplace = True)
df
```

```python
print(df.isnull().sum())
```

```python
# Step 5: Fix Inconsistent Text Formatting

df["Department"] = df["Department"].str.upper()
df
```

```python
# Step 6: Convert Admission Data into standard Datetime Format

df["Admission Date"] = pd.to_datetime(df["Admission Date"], errors = 'coerce')
df

#If a value cannot be converted to a valid date then it will display
#NAT - Not a Time'''
```

```python
# Step 7: Save Cleaned Datasheet

df.to_csv("cleaned_student_data.csv", index = False)

print("\n Cleaned CSV file is saved as cleaned_student_csv")
```
