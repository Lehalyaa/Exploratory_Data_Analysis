```python
# Add a New Column

df["Grade"] = ["Second Class", "Distinction", "First Class", "First Class", "Third Class"]

df
```

```python
# Update Data in DataFrame
# Change marks of first student to 89

df.loc[0, "Marks"] = 89
df
```

```python
# Delete a Column

df.drop("Grade", axis=1, inplace = True)
''' axis = 1 --> column, axis = 0 --> row '''
''' inplace is used to make the changes permanently to the dataframe'''
df
```

```python
# Perform Basic Statistical Analysis

mean1 = df["Marks"].mean()
print("Mean: ", mean1)
```

```python
# Aggregate Functions in Data Frames

min_value = df["Marks"].min()
max_value = df["Marks"].max()
sum_value = df["Marks"].sum()

print("Minimum Value of Marks: ", min_value)
print("Maximum Value of Marks: ", max_value)
print("Sum of Marks: ", sum_value)
```

```python
# Apply conditions (Filtering)
# Display Students who scored more than 90

df[df["Marks"] > 90]
```
